```python
def generate_stops_df(route_list):
    stops_df = pd.DataFrame()
    for route in route_list:
        print(route)
        try:
            # Make the API request
            response = requests.get(url=f"https://bustime.mta.info/api/where/stops-for-route/MTA%20NYCT_{route}.json?key={BUS_TIME_KEY}&includePolylines=false&version=2")
            
            # Check if the response is valid and contains the expected data
            response.raise_for_status()  # This will raise an error for bad responses (e.g., 404, 500)
            stops_data = response.json()
            
            # Check if 'stops' key exists in the response
            if 'data' in stops_data and 'references' in stops_data['data'] and 'stops' in stops_data['data']['references']:
                stops = pd.DataFrame(stops_data['data']['references']['stops'])
                stops['route'] = route
                stops_df = pd.concat([stops_df, stops], ignore_index=True)
            else:
                print(f"No stops data found for route {route}")
        
        except Exception as e:
            # Handle the exception and skip this route
            print(f"Failed to fetch or process data for route {route}: {e}")
            continue  # Continue to the next route if an error occurs
    
    return stops_df
```

```python
def get_gap(stop):
    df = get_stop_data(stop)
    
    if df.empty:
        return pd.DataFrame()
    
    # Check if the required column exists
    if 'MonitoredVehicleJourney.MonitoredCall.ExpectedArrivalTime' not in df.columns:
        return pd.DataFrame()
    
    # Drop rows with missing values in the specified column
    df.dropna(subset=['MonitoredVehicleJourney.MonitoredCall.ExpectedArrivalTime'], inplace=True)
    
    # If the DataFrame is empty after dropping rows, return an empty DataFrame
    if df.empty:
        return pd.DataFrame()
    df['gap'] = pd.to_datetime(df['MonitoredVehicleJourney.MonitoredCall.ExpectedArrivalTime']) - pd.to_datetime(df['MonitoredVehicleJourney.MonitoredCall.AimedArrivalTime'])
    df['gap'] = df['gap'].dt.total_seconds() / 60
    df['absolute_gap'] = df['gap'].abs()
    mean_gap = df['absolute_gap'].mean()
    stop_gap_df = pd.DataFrame({'stop': [stop], 'mean_gap': [mean_gap]})
    return stop_gap_df
```

```python
def gap_for_stops(stop_list,size):
    df = pd.DataFrame()
    
    reduced_stop_list = random.sample(stop_list, size)
    for stop in tqdm(reduced_stop_list):
        try:
            stop_df = get_gap(stop)
            
            # Only concatenate non-empty DataFrames
            if not stop_df.empty:
                df = pd.concat([df, stop_df], ignore_index=True)

        except Exception as e:
            # Log the error and continue to the next stop
            print(f"Error processing stop {stop}: {e}")
            continue

    return df

```

```python
def extract_segment(route, stop, next_stop):
    """
    Extracts the segment of the route between the stop and next_stop.
    """
    if not route or not stop or not next_stop:
        return None
    
    # Ensure route is a LineString or MultiLineString
    line_segments = list(route.geoms) if route.geom_type == 'MultiLineString' else [route]
    
    # Iterate through the segments to find the one containing the points
    for segment in line_segments:
        if segment.distance(stop) < 1e-6 and segment.distance(next_stop) < 1e-6:
            return segment 
            
    
    # If not found, return the shortest line segment between the stops
    return LineString([stop, next_stop])
   
```

```python
def make_slow_routes(df,count,geo_stop_dict,geo_next_stop_dict):
    route_speed = df.groupby(['route_id','timepoint_stop_name','next_timepoint_stop_name'])['average_road_speed'].mean().reset_index()
    slowest_routes = route_speed.sort_values('average_road_speed',ascending=True).head(count)

    slowest_routes['stop_geo'] = slowest_routes['timepoint_stop_name'].map(geo_stop_dict)
    slowest_routes['next_stop_geo'] = slowest_routes['next_timepoint_stop_name'].map(geo_next_stop_dict)

    return slowest_routes
```

```python
def make_gpd(df,bus_routes):
    df = df.merge(bus_routes, left_on='route_id', right_on='route_id',how='left')
    df['stop_point'] = df['stop_geo'].apply(lambda x: Point(x['coordinates']))
    df['next_stop_point'] = df['next_stop_geo'].apply(lambda x: Point(x['coordinates']))
    
    df['segment_geometry'] = df.apply(lambda row: extract_segment(row['geometry'], row['stop_point'], row['next_stop_point']), axis=1)
    short_df = df[['route_id','timepoint_stop_name','next_timepoint_stop_name','average_road_speed','segment_geometry']]
    gdf = gpd.GeoDataFrame(short_df, geometry='segment_geometry')
    return gdf
    
```

```python
def make_table(df):
    boroughs = df.groupby('borough')['average_road_speed'].mean().reset_index(name='avg_speed')
    whole_city = df['average_road_speed'].mean()
    table = pd.DataFrame({'borough': 'NYC Whole', 'avg_speed': whole_city}, index=[0])
    boroughs = pd.concat([boroughs, table], ignore_index=True)
    boroughs['avg_speed'] = boroughs['avg_speed'].round(1)
    return boroughs
```
